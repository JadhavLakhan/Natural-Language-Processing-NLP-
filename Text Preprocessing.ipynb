```python
#Import Library
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
# Tokenization :- Converts paragraph sentences into words
nltk.download('punkt_tab')  # Download the 'punkt_tab' resource instead of 'punkt'
sentences = nltk.sent_tokenize(Paragraph)
```

```python
# Tokenization :- Converts paragraph  sentences into words
nltk.download('punkt')
sentences=nltk.sent_tokenize(Paragraph)
```

```python
Stemmer=PorterStemmer()
```

```python
Stemmer.stem('statistics')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
import re
```

```python

corpus=[]
for i in range(len(sentences)):
  text=re.sub('[^a-zA-Z]',' ',sentences[i])
  text=text.lower()
  corpus.append(text)
```

```python
corpus
```

```python
stopwords.words('english')
```

```python
## stemming
for i in corpus:
  words=nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(Stemmer.stem(word))
```

```python
## Lemmatization
for i in corpus:
  words=nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(lemmatizer.lemmatize(word))
```

```python
## Apply Stopwords .Lemmatization
corpus=[]
for i in range(len(sentences)):
  text=re.sub('[^a-zA-Z]',' ',sentences[i])
  text=text.lower()
  text=text.split()
  text=[lemmatizer.lemmatize(word) for word in text if word not in set(stopwords.words('english'))]
  text=' '.join(text)
  corpus.append(text)
```

```python
## Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(binary=True)
```

```python
X=cv.fit_transform(corpus)
```
