```python
import gensim
```

```python
from gensim.models import Word2Vec,KeyedVectors
```

```python
import gensim.downloader as api
wv=api.load('word2vec-google-news-300')
vec_king=wv['king']
```

```python
vec_king #300 dimensions
```

```python
wv['cricket']
```

```python
wv.most_similar('cricket') #most similar vector (cosine similarity)

#Values range from -1 to 1:
#1: Perfectly similar (identical context/meaning).
#0: No similarity.
#-1: Opposite meanings.
```

```python
wv.most_similar(negative=['king'])
```

```python
wv.similarity('man','king')
```

```python
wv.similarity('man','cricket')
```

```python
wv.similarity('Mathematics','statistic')
```

```python
vec=wv['king']-wv['man']+wv['woman']
```

```python
wv.most_similar([vec])
```
